```python

"""
#### 和法向量内积相同的所有点构成一条直线
> 二维点构成直线
> 三维点构成平面
> N维点构成超平面

内积是投影乘棍长(法向量的模长)
投影是内积比棍长(法向量的模长)


In geometry, a normal is an object such as a line, ray, or vector that is perpendicular to a given object. 
perpendicular [< L perpendiculāris < perpendiculum plumbline < PER-+ pendēre to hang, weigh]

Things Happen for A Reason
"""
                                                                                                                                                                  
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

point  = np.array([1, 2])       
normal = np.array([0.1, -0.2])  # 法向量

# a plane is a*x+b*y+c*z+d=0
# [a,b,c] is the normal. Thus, we have to calculate
# d and we're set
d = -point.dot(normal)

# create x,y
xx, yy = np.meshgrid(range(10), range(10))

# calculate corresponding z
z = (-normal[0] * xx - normal[1] * yy - d) * 1. /normal[2]

# plot the surface
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(xx, yy, z)
plt.show()
```
